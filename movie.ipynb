```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
      l.append(i['name'])
      return l
```

```python
movies['genres']=movies['genres'].apply(convert) 
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter +=1 
        else:
            break 
        return l 
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l 
```

```python
movies.loc[:,'crew']=movies['crew'].apply(fetch_director) 
```

```python
movies.loc[:,'genres']=movies['genres'].apply(lambda X: [i.replace (" ","") for i in X]if isinstance(X,list)else [])
movies.loc[:,'keywords']=movies['keywords'].apply(lambda X:[i.replace(" ","") for  i in X]if isinstance(X,list)else[]) 
movies.loc[:,'cast']=movies['cast'].apply(lambda X: [i.replace(" ","")for i in X]if isinstance (X,list)else[])
movies.loc[:,'crew']=movies['crew'].apply(lambda X: [i.replace(" ","")for i in X]if isinstance(X,list)else [])
```

```python
from  nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[ ]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```
